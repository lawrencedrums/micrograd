````python
def demo():
    h = 0.0001
    
    a = Value(2.0)   ; a.label = 'a'
    
    w1 = Value(-3.0)  ; w1.label = 'w1'
    w2 = Value(8.0)   ; w2.label = 'w2'
    
    w1w2 = w1 + w2    ; w1w2.label = 'w1w2'
    
    L = a * w1w2       ; L.label = 'L' ; L.grad = 1.0

    return L
````

````python
draw_graph(L)
````

We have
`L = a * w1w2`

We want to know
`dL/da =? w1w2`

```
(f(a + h) - f(a)) / h
((a*f + h*f) - a*f) / h
h*f / h
f

Therefore the grad of L w.r.t a is just w1w2
```

We have
`w1w2 = w1 + w2`

We want to know
`dL/dw2 =? w1`

```
(f(x + h) - f(x)) / h
((w1 + h + w2) - (w1 + w2)) / h
h / h
1

Therefore the grad of L w.r.t w2 is just 1, and also 1 w.r.t w1.
```
